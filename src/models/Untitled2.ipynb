```python
import pandas as pd
```

```python
# create a dataframe of dates
df = pd.DataFrame({'Dates': pd.date_range("2019-01-01", "2021-04-27")})

# create a series of business days
busines_dates = pd.bdate_range("2019-01-01", "2021-04-27")

# find where the two do not intersect
weekends = df.loc[~df['Dates'].isin(busines_dates)].Dates.dt.strftime('%Y-%m-%d').to_list()
```
